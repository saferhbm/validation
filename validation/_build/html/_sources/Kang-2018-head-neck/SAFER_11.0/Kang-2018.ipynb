> - ***Setup***                    :Cervical spine FSU 
>   * `MAT_RIGID` sled with material properties of steel.
>   * `MAT_RIGID` elliptical ring with weight $5 kg$ to match the pretension forces applied during tests. Elliptical ring with major diameter=$311 mm$ and  minor diameter=$172 mm$
>   * `MAT_ELASTIC` potting material with $E=03.42Gpa$, $density=1.14 g/cm^3$ and $pr=0.33$
> - ***Contacts***                        : `*CONTACT_AUTOMATIC_SURFACE_TO_SURFACE` with $fs=0.3 , vdc=20, soft=2$ between the HBM and the elliptical ring
> ```{admonition} Positioning and Pre-simulation
>:class: important 
>
> - **Initial position - specimen**:
>   * Horizontal Distance from head CoG to center axis of the potting is defined as $81 \pm 34 mm$
>   * Vertical Distance from head CoG to center axis of the potting is defined as $229 \pm 21 mm$
>   * angle of the T1 superior endplate maintained during porring around $24 \pm 5 degree$
> - **Initial position - Potting w.r.t the elliptical ring**:
>   * potting center to elleptical ring center is $28 mm$ apart about the minor axis of the elliptical ring.
>   * Vertical Distance from head CoG to center axis of the potting is defined as $229 \pm 21 mm$
> - **Pre-Simulation**:
>   * The specimen is let to settle under gravity for the neck muscles to get pretensed for $300 ms$ with defined `GLOBAL_DAMPING` of magnitude $0.057$
>```
> - **Constraints** 
>   * `*CONSTRAINT_SHELL_IN_SOLID_PENALTY` used for the potting. 
>   * `*BOUNDARY_PRESCRIBED_MOTION_RIGID` to set the fix the Dof of the elleptical ring and seld other than the loading direction.
> ```{important} 
> The elliptical ring is free to move vertically during presimulation.
>```
> - **Loads**            : 
>   * `*BOUNDARY_PRESCRIBED_MOTION_RIGID` define constant acceleration pulse in frontal, side and oblique angles to the sled and ellecptical ring.

### Side Impact

````python
exp_side=pd.read_csv(os.path.join(experiment_dir,'side.csv'),delimiter=';',header=[0,1,2])
measures_side=['ax','az','rotx','Mx','Fy','Fz']
## Resample data
def resam(xdata,ydata,sTime,endtime):
    x = xdata
    y = ydata
    # Interpolate the data to the desired length
    newx = np.arange(sTime,endtime,0.1)#10kHz resolution
    newy = np.interp(newx, x, y)
    return np.array(newy).flatten()
T_samp=np.arange(0,200,0.1)
test_average_side={}
for measure in measures_side:
    df=pd.DataFrame(T_samp)
    for pmhs in range(1,3):
        ref_t=exp_side[measure]['{}'.format(pmhs)].X.dropna()
        ref_val=exp_side[measure]['{}'.format(pmhs)].Y.dropna()
        r_v=resam(ref_t,ref_val,0,200)
        df[pmhs]=r_v
    test_average_side[measure]=df.iloc[:,1:].mean(axis=1)
Sim_ISO={}

peak_vals_side={}
peak120_vals_side={}
p_120_index_side={}
for m in measures_side:
    peak_vals_side[m]=T_samp[test_average_side[m][abs(test_average_side[m])==max(abs(test_average_side[m]))].index[0]]
    peak120_vals_side[m]=T_samp[test_average_side[m][abs(test_average_side[m])==max(abs(test_average_side[m]))].index[0]]*1.2
    p_120_index_side[m]=np.where(T_samp<peak120_vals_side[m])[0][-1]
````
